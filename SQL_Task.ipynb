Database & Table Operations (SQLite)**bold text**

Prompt:*Write a Python program using SQLite in Google Colab that demonstrates the following SQL operations:*
*   Create a database connection
*   Create an Employee table
*   Display the table schema
*   Verify the table has been created



```python
import sqlite3
import pandas as pd

# Create Database Connection
conn = sqlite3.connect("company.db")

# Create Cursor
cursor = conn.cursor()

# Create Employee Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee(
    EmployeeID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Department TEXT,
    Salary REAL,
    Experience INTEGER
)
""")

print("Employee table created successfully!")

# Display Table Schema
print("\nEmployee Table Schema:")

schema = pd.read_sql_query(
    "PRAGMA table_info(Employee);",
    conn
)

print(schema)

# Verify Table Exists
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)

print("\nAvailable Tables:")
print(tables)

# Close Connection
conn.close()
```

**INSERT, UPDATE & DELETE Operations**

Prompt:*Write a Python program using SQLite that demonstrates the following SQL operations:*
*   Connect to the database
*   Insert single and multiple records into the Employee table
*   Display all records
*   Update an employee's salary
*   Delete an employee record
*  Display the updated table

```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Clear existing records (Optional)
cursor.execute("DELETE FROM Employee")

# Insert a Single Record
cursor.execute("""
INSERT INTO Employee
(EmployeeID, Name, Department, Salary, Experience)
VALUES
(101, 'Alice', 'IT', 50000, 3)
""")

# Insert Multiple Records
employees = [
    (102, 'Bob', 'HR', 45000, 2),
    (103, 'Charlie', 'Finance', 60000, 5),
    (104, 'David', 'IT', 55000, 4),
    (105, 'Eva', 'Marketing', 48000, 2)
]

cursor.executemany("""
INSERT INTO Employee
VALUES (?, ?, ?, ?, ?)
""", employees)

conn.commit()

# Display Records
print("Employee Table")
print(pd.read_sql_query(
    "SELECT * FROM Employee",
    conn
))

# Update Salary
cursor.execute("""
UPDATE Employee
SET Salary = 65000
WHERE EmployeeID = 103
""")

conn.commit()

print("\nAfter Updating Charlie's Salary")
print(pd.read_sql_query(
    "SELECT * FROM Employee",
    conn
))

# Delete Employee
cursor.execute("""
DELETE FROM Employee
WHERE EmployeeID = 102
""")

conn.commit()

print("\nAfter Deleting Bob")
print(pd.read_sql_query(
    "SELECT * FROM Employee",
    conn
))

conn.close()
```

*   Retrieve records using IN

```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")

print("1. Display All Records")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

print("\n2. Display Specific Columns")
print(pd.read_sql_query(
    "SELECT Name, Department, Salary FROM Employee;",
    conn
))

print("\n3. WHERE Clause (Department = 'IT')")
print(pd.read_sql_query(
    "SELECT * FROM Employee WHERE Department='IT';",
    conn
))

print("\n4. AND Condition")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    WHERE Department='IT'
    AND Salary > 50000;
    """,
    conn
))

print("\n5. OR Condition")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    WHERE Department='IT'
    OR Department='Finance';
    """,
    conn
))

print("\n6. IN Operator")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    WHERE Department IN ('IT','HR');
    """,
    conn
))

print("\n7. BETWEEN Operator")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    WHERE Salary BETWEEN 50000 AND 65000;
    """,
    conn
))

print("\n8. LIKE Operator")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    WHERE Name LIKE 'A%';
    """,
    conn
))

print("\n9. ORDER BY Salary DESC")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    ORDER BY Salary DESC;
    """,
    conn
))

print("\n10. LIMIT")
print(pd.read_sql_query(
    """
    SELECT * FROM Employee
    LIMIT 3;
    """,
    conn
))

conn.close()
```

**SQL Aggregate Functions & Grouping**

Prompt:*Write a Python program using SQLite that demonstrates the following SQL Aggregate Functions and Grouping operations:*
*   Display unique departments using DISTINCT
*   Count the total number of employees
*   Calculate the total salary using SUM()
*   Find the highest salary using MAX()
*   Find the lowest salary using MIN()
*   Group employees by department using GROUP BY
*   Filter grouped results using HAVING

```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")

# DISTINCT
print("1. DISTINCT Departments")
print(pd.read_sql_query("""
SELECT DISTINCT Department
FROM Employee;
""", conn))

# COUNT
print("\n2. Total Employees")
print(pd.read_sql_query("""
SELECT COUNT(*) AS Total_Employees
FROM Employee;
""", conn))

# SUM
print("\n3. Total Salary")
print(pd.read_sql_query("""
SELECT SUM(Salary) AS Total_Salary
FROM Employee;
""", conn))

# AVG
print("\n4. Average Salary")
print(pd.read_sql_query("""
SELECT AVG(Salary) AS Average_Salary
FROM Employee;
""", conn))

# MAX
print("\n5. Highest Salary")
print(pd.read_sql_query("""
SELECT MAX(Salary) AS Highest_Salary
FROM Employee;
""", conn))

# MIN
print("\n6. Lowest Salary")
print(pd.read_sql_query("""
SELECT MIN(Salary) AS Lowest_Salary
FROM Employee;
""", conn))

# GROUP BY
print("\n7. Average Salary by Department")
print(pd.read_sql_query("""
SELECT Department,
COUNT(*) AS Employees,
AVG(Salary) AS Average_Salary
FROM Employee
GROUP BY Department;
""", conn))

# HAVING
print("\n8. Departments with Average Salary > 50000")
print(pd.read_sql_query("""
SELECT Department,
AVG(Salary) AS Average_Salary
FROM Employee
GROUP BY Department
HAVING AVG(Salary) > 50000;
""", conn))

conn.close()
```

**ALTER TABLE & Table Management**

Prompt:*Write a Python program using SQLite that demonstrates the following SQL table management operations:*
*   Display the existing Employee table

*   Update the values of the new column
*   Rename an existing column
*   Rename the table
*   Display the modified table
*   Drop the table
*   Verify that the table has been deleted



```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Display Existing Table
print("1. Employee Table")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

# Add New Column
cursor.execute("""
ALTER TABLE Employee
ADD COLUMN City TEXT;
""")

# Update New Column
cursor.execute("""
UPDATE Employee
SET City = 'Hyderabad'
WHERE Department = 'IT';
""")

cursor.execute("""
UPDATE Employee
SET City = 'Bangalore'
WHERE Department = 'Finance';
""")

cursor.execute("""
UPDATE Employee
SET City = 'Chennai'
WHERE Department = 'Marketing';
""")

conn.commit()

print("\n2. After Adding City Column")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

# Rename Column
cursor.execute("""
ALTER TABLE Employee
RENAME COLUMN Salary TO MonthlySalary;
""")

conn.commit()

print("\n3. After Renaming Salary Column")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

# Rename Table
cursor.execute("""
ALTER TABLE Employee
RENAME TO EmployeeDetails;
""")

conn.commit()

print("\n4. Tables in Database")
print(pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
))

print("\n5. EmployeeDetails Table")
print(pd.read_sql_query(
    "SELECT * FROM EmployeeDetails;",
    conn
))

# Drop Table
cursor.execute("""
DROP TABLE EmployeeDetails;
""")

conn.commit()

print("\n6. After Dropping Table")
print(pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
))

conn.close()
```

**SQL JOIN Operations**

Prompt:*Write a Python program using SQLite that demonstrates the following SQL JOIN operations:*
*   Create Employee and Department tables
*   Insert records into both tables
*   Display both tables
*   Perform INNER JOIN
*   Perform LEFT JOIN
*   Perform CROSS JOIN



```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop tables if they already exist
cursor.execute("DROP TABLE IF EXISTS Employee")
cursor.execute("DROP TABLE IF EXISTS Department")

# Create Employee Table
cursor.execute("""
CREATE TABLE Employee(
    EmployeeID INTEGER PRIMARY KEY,
    Name TEXT,
    DepartmentID INTEGER,
    Salary REAL
)
""")

# Create Department Table
cursor.execute("""
CREATE TABLE Department(
    DepartmentID INTEGER PRIMARY KEY,
    DepartmentName TEXT,
    Manager TEXT
)
""")

# Insert Employee Records
employees = [
    (101, "Alice", 1, 50000),
    (102, "Bob", 2, 45000),
    (103, "Charlie", 1, 65000),
    (104, "David", 3, 55000),
    (105, "Eva", 2, 48000)
]

cursor.executemany(
    "INSERT INTO Employee VALUES (?, ?, ?, ?)",
    employees
)

# Insert Department Records
departments = [
    (1, "IT", "Robert"),
    (2, "HR", "Sophia"),
    (3, "Finance", "William")
]

cursor.executemany(
    "INSERT INTO Department VALUES (?, ?, ?)",
    departments
)

conn.commit()

# Display Employee Table
print("Employee Table")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

# Display Department Table
print("\nDepartment Table")
print(pd.read_sql_query(
    "SELECT * FROM Department;",
    conn
))

# INNER JOIN
print("\nINNER JOIN")
print(pd.read_sql_query("""
SELECT
Employee.EmployeeID,
Employee.Name,
Department.DepartmentName,
Department.Manager,
Employee.Salary
FROM Employee
INNER JOIN Department
ON Employee.DepartmentID =
Department.DepartmentID;
""", conn))

# LEFT JOIN
print("\nLEFT JOIN")
print(pd.read_sql_query("""
SELECT
Employee.Name,
Department.DepartmentName,
Employee.Salary
FROM Employee
LEFT JOIN Department
ON Employee.DepartmentID =
Department.DepartmentID;
""", conn))

# CROSS JOIN
print("\nCROSS JOIN")
print(pd.read_sql_query("""
SELECT
Employee.Name,
Department.DepartmentName
FROM Employee
CROSS JOIN Department;
""", conn))

conn.close()
```

**SQL Subqueries & Views**

Prompt:*Write a Python program using SQLite that demonstrates the following SQL operations:*
*   Display the Employee table
*   Find employees whose salary is greater than the average salary using a Subquery
*   Find employees working in the IT department using an IN Subquery
*   Check employee existence using EXISTS
*   Create a SQL View
*   Display records from the View
*   Drop the View
*   Verify that the View has been deleted

```python
import sqlite3
import pandas as pd

# Connect to Database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Display Employee Table
print("Employee Table")
print(pd.read_sql_query(
    "SELECT * FROM Employee;",
    conn
))

# ----------------------------
# Scalar Subquery
# ----------------------------
print("\nEmployees with Salary Greater than Average Salary")

print(pd.read_sql_query("""
SELECT *
FROM Employee
WHERE Salary >
(
    SELECT AVG(Salary)
    FROM Employee
);
""", conn))

# ----------------------------
# IN Subquery
# ----------------------------
print("\nEmployees Working in IT Department")

print(pd.read_sql_query("""
SELECT *
FROM Employee
WHERE DepartmentID IN
(
    SELECT DepartmentID
    FROM Department
    WHERE DepartmentName='IT'
);
""", conn))

# ----------------------------
# EXISTS
# ----------------------------
print("\nUsing EXISTS")

print(pd.read_sql_query("""
SELECT Name
FROM Employee
WHERE EXISTS
(
    SELECT *
    FROM Department
    WHERE Employee.DepartmentID =
          Department.DepartmentID
);
""", conn))

# ----------------------------
# Create View
# ----------------------------
cursor.execute("""
CREATE VIEW IF NOT EXISTS Employee_View AS

SELECT
Employee.Name,
Department.DepartmentName,
Employee.Salary

FROM Employee

INNER JOIN Department

ON Employee.DepartmentID =
Department.DepartmentID;
""")

conn.commit()

print("\nEmployee View")

print(pd.read_sql_query("""
SELECT *
FROM Employee_View;
""", conn))

# ----------------------------
# Drop View
# ----------------------------
cursor.execute("""
DROP VIEW Employee_View;
""")

conn.commit()

print("\nAvailable Views")

print(pd.read_sql_query("""
SELECT name
FROM sqlite_master
WHERE type='view';
""", conn))

conn.close()
```

**End-to-End SQL Project**

Prompt:*Write a Python program using SQLite that demonstrates a complete SQL workflow by performing the following operations:*
*   Create Customer and Orders tables
*   Insert records into both tables
*   Display the tables
*   Update a customer record
*   Delete an order
*   Display unique customer cities
*   Calculate total, average, minimum and maximum order amount
*   Group orders by customer
*   Filter grouped data using HAVING
*   Perform an INNER JOIN
*   Perform a LEFT JOIN
*   Find customers with orders greater than the average amount using a subquery
*   Create and display a View
*   Drop the View

```python
import sqlite3
import pandas as pd

# Create Database
conn = sqlite3.connect("sales.db")
cursor = conn.cursor()

# Drop Existing Tables
cursor.execute("DROP TABLE IF EXISTS Orders")
cursor.execute("DROP TABLE IF EXISTS Customers")

# Create Customers Table
cursor.execute("""
CREATE TABLE Customers(
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    City TEXT
)
""")

# Create Orders Table
cursor.execute("""
CREATE TABLE Orders(
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    Product TEXT,
    Amount REAL,
    FOREIGN KEY(CustomerID)
    REFERENCES Customers(CustomerID)
)
""")

# Insert Customer Records
customers = [
    (1,"Alice","Hyderabad"),
    (2,"Bob","Chennai"),
    (3,"Charlie","Bangalore"),
    (4,"David","Hyderabad")
]

cursor.executemany(
    "INSERT INTO Customers VALUES(?,?,?)",
    customers
)

# Insert Order Records
orders = [
    (101,1,"Laptop",65000),
    (102,2,"Mouse",1000),
    (103,1,"Keyboard",2500),
    (104,3,"Monitor",12000),
    (105,4,"Printer",8000)
]

cursor.executemany(
    "INSERT INTO Orders VALUES(?,?,?,?)",
    orders
)

conn.commit()

# Display Customers
print("Customers")
print(pd.read_sql_query(
    "SELECT * FROM Customers",
    conn
))

# Display Orders
print("\nOrders")
print(pd.read_sql_query(
    "SELECT * FROM Orders",
    conn
))

# Update
cursor.execute("""
UPDATE Customers
SET City='Pune'
WHERE CustomerID=2
""")

conn.commit()

# Delete
cursor.execute("""
DELETE FROM Orders
WHERE OrderID=102
""")

conn.commit()

# DISTINCT
print("\nUnique Cities")
print(pd.read_sql_query("""
SELECT DISTINCT City
FROM Customers
""", conn))

# Aggregate Functions
print("\nOrder Statistics")
print(pd.read_sql_query("""
SELECT
COUNT(*) AS TotalOrders,
SUM(Amount) AS TotalAmount,
AVG(Amount) AS AverageAmount,
MIN(Amount) AS MinimumAmount,
MAX(Amount) AS MaximumAmount
FROM Orders
""", conn))

# GROUP BY
print("\nOrders by Customer")
print(pd.read_sql_query("""
SELECT
CustomerID,
COUNT(*) AS TotalOrders,
SUM(Amount) AS TotalPurchase
FROM Orders
GROUP BY CustomerID
""", conn))

# HAVING
print("\nCustomers Spending More Than 10000")
print(pd.read_sql_query("""
SELECT
CustomerID,
SUM(Amount) AS TotalPurchase
FROM Orders
GROUP BY CustomerID
HAVING SUM(Amount)>10000
""", conn))

# INNER JOIN
print("\nINNER JOIN")
print(pd.read_sql_query("""
SELECT
Customers.CustomerName,
Orders.Product,
Orders.Amount
FROM Customers
INNER JOIN Orders
ON Customers.CustomerID=Orders.CustomerID
""", conn))

# LEFT JOIN
print("\nLEFT JOIN")
print(pd.read_sql_query("""
SELECT
Customers.CustomerName,
Orders.Product,
Orders.Amount
FROM Customers
LEFT JOIN Orders
ON Customers.CustomerID=Orders.CustomerID
""", conn))

# Subquery
print("\nOrders Above Average Amount")
print(pd.read_sql_query("""
SELECT *
FROM Orders
WHERE Amount >
(
SELECT AVG(Amount)
FROM Orders
)
""", conn))

# Create View
cursor.execute("""
CREATE VIEW CustomerOrders AS

SELECT
Customers.CustomerName,
Orders.Product,
Orders.Amount

FROM Customers

INNER JOIN Orders

ON Customers.CustomerID=Orders.CustomerID
""")

conn.commit()

print("\nCustomerOrders View")
print(pd.read_sql_query(
    "SELECT * FROM CustomerOrders",
    conn
))

# Drop View
cursor.execute("DROP VIEW CustomerOrders")

conn.commit()

print("\nView Dropped Successfully")

conn.close()
```
